```python
import datetime

pisos = 10
departamentos_por_piso = 4
departamentos_vendidos = []
departamentos_precios_v = [3800 , 3000, 2800, 3500]
compradores = []



def mostrar_menu():
    print("\n*______*_MENÚ_*______*\n")
    print("1. Comprar Departamento")
    print("2. Mostrar Departamentos Disponibles")
    print("3. Ver Listado de Compradores")
    print("4. Mostrar Ganancias Totales")
    print("5. Salir")
    opcion = input("Elija una opción:")
    return opcion


def comprar_departamento():
    print("\n*______*_COMPRAR DEPARTAMENTO_*______\n")
    piso = int(input("Ingrese número de piso (1 al 10): ".format(pisos)))
    print("---------------------------------")
    print("| Tipo Departamento |  Valores  |")
    print("|      A            |  3.800 UF |")
    print("|      B            |  3.000 UF |")
    print("|      C            |  2.800 UF |")
    print("|      D            |  3.500 UF |")
    print("=================================")
    tipo_departamento = input("Ingrese tipo de departamento (A, B, C, D ): ")
    departamento = tipo_departamento + str(piso)

    if departamento in departamentos_vendidos:
        print("Departamento NO Disponible!")
    else:
        print("Departamento DISPONIBLE!")
        print("1. Comprar")
        opcion = int(input("Elija una opción: "))
        run = input("Ingrese el RUT del Comprador: ")

        if opcion == 1:
            departamentos_vendidos.append(departamento)
            compradores.append(run)
            print("COMPRA SE REALIZÓ EXITOSAMENTE")
            print("MUCHAS FELICIDADES POR TU NUEVO HOGAR")



def mostrar_departamentos_disponibles():
    print("\*_____*_DEPARTAMENTOS DISPONIBLES_*______*\n")
    print("---------------------------------------------")
    print("| PISO   |        TIPO   |")
    print("---------| A | B | C | D |")
    print("|  1     |   |   |   |   |")
    print("|  2     |   |   |   |   |")
    print("|  3     |   |   |   |   |")
    print("|  4     |   |   |   |   |")
    print("|  5     |   |   |   |   |")
    print("|  6     |   |   |   |   |")
    print("|  7     |   |   |   |   |")
    print("|  8     |   |   |   |   |")
    print("|  9     |   |   |   |   |")
    print("|  10    |   |   |   |   |")
    print("--------------------------")
    print("Departamentos vendidos: ", departamentos_vendidos)
    print("Departamentos Disponibles:")
    for piso in range(1, pisos+1):
        for tipo_departamento in ["A", "B", "C", "D"]:
            departamento = tipo_departamento + str(piso)
            if departamento not in departamentos_vendidos:
                print("DEPARTAMENTO DISPONIBLE")


def mostrar_listado_compradores():
    print("\n*______*_LISTA DE COMPRADORES_*______*\n")
    compradores_ordenados = sorted(compradores)
    for comprador in compradores_ordenados:
        print("COMPRADOR")


def mostrar_ganancias_totales():
    print("\n*______*_GANACIAS TOTALES_*______*\n")
    ganancias_vendidos = sum([departamentos_precios_v[pisos.index(depto[1])] for depto in departamentos_vendidos])
    ganancias_totales = ganancias_vendidos
    print("Ganancias por ventas: ", ganancias_vendidos)
    print("Ganancias totales: ", ganancias_totales)


def ejecucion_practica():
    print("====_Bienvenidos a Inmobiliaria Casa Feliz_====")
    tiempo_inicio = datetime.datetime.now()

    while  True :
        opcion = mostrar_menu()

        if opcion == '1':
            comprar_departamento()
        elif opcion == '2':
            mostrar_departamentos_disponibles()
        elif opcion == '3':
            mostrar_listado_compradores()
        elif opcion == '4':
            mostrar_ganancias_totales()
        elif opcion == '5':
            tiempo_fin = datetime.datetime.now()
            duracion = tiempo_fin - tiempo_inicio
            print("Gracias por preferir Inmobiliaria Casa Feliz!!!")
            print("Disfruta tu nuevo Hogar")
            print("Duración de la ejecución: ", duracion)
            break
        else:
            print("OPCIÓN NO VALIDA!, Inténtelo nuevamente porfavor!")


ejecucion_practica()
```
